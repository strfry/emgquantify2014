```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sys
import os

from mpltools import style
style.use('mystyle')

from HDF5File import HDF5File

path = './Data/'
#filename = 'florian_10kg_stehen_2014.12.19.hdf5'
#filename = 'tobias_5kg_2014.12.15.hdf5'
filename = 'jonathan_7.5kg_rechts_2014.14.12.hdf5'

# Open HDF5 File
f = HDF5File(path + filename)
```

# Compute the RMS (Root Mean Square) of each trial

```python
def quad(x): return x**2

def rms(samples):
    squares = map(quad, samples)
    s = sum(squares) / len(samples)
    return np.sqrt(s)

sums = map(rms, f.trials)

plt.plot(sums)
```

# Combine RMS with trial length

```python
def transform(trial):
    length = len(trial) / 256.0
    power = rms(trial)
    
    return power, length

powers, lengths = zip(*map(transform, f.trials))

fig, ax1 = plt.subplots()

ax1.plot(powers)
ax1.set_ylabel('RMS', color='r')

ax2 = ax1.twinx()
ax2.plot(lengths, color='b')
ax2.set_ylabel('Time (s)', color="b")
```

Lets further analyze the power aspect in the frequency domain.

In some cases we notices a peak below 5 Hz. We check if ignoring frequencies below 5 Hz makes a difference:

```python
LOW_CUTOFF = 5

def transform(trial):
    data = trial
    
    data = np.hanning(len(trial)) * data
    
    F = np.fft.rfft(data)
    F2 = F[LOW_CUTOFF:]
    
    sum1 = rms(abs(F))
    sum1 = sum1 / np.sqrt(len(F))
    
    sum2 = rms(abs(F2))
    sum2 = sum2 / np.sqrt(len(F2))
    
    return sum1, sum2

sum1, sum2 = zip(*map(transform, f.trials))

plt.plot(sum1)
plt.plot(sum2)
```
